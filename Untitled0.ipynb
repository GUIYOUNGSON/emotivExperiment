```python
import os
import shutil
prefix = './frontend/imagesCopy/indoor/'
dirlist = os.listdir(prefix)
print len(dirlist)
i = 0
for file in dirlist:
    shutil.copy2(prefix + file, './frontend/imagesRenamed/indoor/' 
                 + str(i) + ".jpg")
    print "Copied " + file + " to " + str(i) + ".jpg"
    i+= 1
```
