```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.decomposition import FastICA, PCA
from scipy import fft
%matplotlib inline
```

```python
f = open("./testdata/journal_7_2014.12.05.03.20")
ar = [line for line in f]
ar = ar[ar.index("-endheader-\n")+1:]
# faces are positive 1, places are -1
trialType = []
trialStart = []
trialEnd = []
DUR = 5000
stimOnsetIdx = len("StimOnset:")
i = 0
for line in ar:
    if "Epoch " in line:
        if "faces" in line:
            trialType.append(1)
        else:
            assert ("places" in line)
            trialType.append(-1)
    elif "Trial:0" in line:
        var = long(line.split(',')[2][stimOnsetIdx:])
        trialStart.append(var)
        trialEnd.append(var + DUR)
        
f = open("./testdata/eeg_7_2014.12.05.03.20")
eegdata = []
labels = []
i = 0
# Make the labels
for line in f:
    if len(line) > 1:
        datapoints = line.split(",")
        timestamp = long(datapoints[-1])
        if timestamp > trialStart[i] and timestamp < trialEnd[i]:
            labels.append(trialType[i])
        elif timestamp > trialEnd[i]:
            i+= 1
            labels.append(trialType[i])
        eegdata.append([float(data) for data in datapoints[:-1]])

eegArray = np.array(eegdata)
eegArray = eegArray[:,3:17]
eegArray = eegArray.transpose()
f, arrs = plt.subplots(14, sharex = True)
i = 0
fig = plt.gcf()
fig.set_size_inches(15,8)
for data in eegArray:
    arrs[i].plot(eegArray[i])
    i += 1

arrs[0].set_title("Raw Channels")
plt.savefig("Raw_Channels")
```

```python
ica = FastICA()
S_ = ica.fit_transform(eegArray.T)  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix
assert np.allclose(eegArray.T, np.dot(S_, A_.T) + ica.mean_)
assert np.allclose(S_, ica.transform(eegArray.T))
```
